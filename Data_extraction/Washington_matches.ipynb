```python
def chunks(lst, n):
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
        

GAMES_TOT = []
for i, u in enumerate ( urls ) :
    r = requests.get( u )
    soup = BeautifulSoup( r.content, 'html.parser' )
    f_all_games = soup.find_all("td", {'class':'score' } )
    games = []

    for s in f_all_games[5:155] :
        games.append( s.get_text() )
        listina = list( chunks( games, 5) )
    GAMES_TOT.append( listina )

#print( GAMES_TOT  )

     
range_GAMES_TOT = list( range(0, len( GAMES_TOT) ) )
print( range_GAMES_TOT )

range_interno_1 = list( range( 0, len( GAMES_TOT[0]), 2 ))
print( range_interno_1 )

range_interno_2 = list( range( 1, len( GAMES_TOT[0]), 2 ))
print( range_interno_2 )

games_1 = []
games_2 = []
for rgt in range_GAMES_TOT:
    for ri1 in range_interno_1:
        games_1.append( GAMES_TOT[rgt][ri1] )
    for ri2 in range_interno_2:
        games_2.append( GAMES_TOT[rgt][ri2] )
            
#print( games_1 )
#print( games_2 )

data_games_1 = pd.DataFrame( games_1 )
data_games_1

data_games_2 = pd.DataFrame( games_2 )
data_games_2
```
