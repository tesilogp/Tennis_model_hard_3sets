```python
def diff_somm( lista_raccolta_1, lista_raccolta_2, lista_diff, lista_plus ) :
    for idx, v in enumerate ( lista_raccolta_1 ) :
        diff = v - lista_raccolta_2[idx]
        plus = v + lista_raccolta_2[idx]
        lista_diff.append( diff )
        lista_plus.append( plus )
```

```python
def desc_master_1000( lista_partite, file_slam_data, lista_raccolta_1,
                      lista_raccolta_2, anni_prima, LABEL_PLACE,
                      lista_diff, lista_plus ) :
    
    ricerca_stringa_in_dataframe_name1( lista_partite, file_slam_data, lista_raccolta_1, anni_prima, LABEL_PLACE )
    ricerca_stringa_in_dataframe_name2( lista_partite, file_slam_data, lista_raccolta_2, anni_prima, LABEL_PLACE )
    diff_somm( lista_raccolta_1, lista_raccolta_2, lista_diff, lista_plus )
```
