```python
import pandas as pd
import numpy as np
import datetime
from datetime import datetime




file_partite = 'file_partite_data_parsed_games.xlsx'
file_partite_read = pd.read_excel( file_partite )
file_partite_data = pd.DataFrame( file_partite_read )
file_partite_data = file_partite_data.sort_values( by='unicode' )
file_partite_data = file_partite_data.loc[ file_partite_data.data_parsed>'2005-01-01' ]
file_partite_data  
```

```python
file_SLAMS = 'file_partite_data_parsed_CLAY.xlsx'
file_SLAMS_read = pd.read_excel( file_SLAMS )
file_SLAMS_data = pd.DataFrame( file_SLAMS_read )
file_SLAMS_data = file_SLAMS_data.loc[ file_SLAMS_data.segno==1 ]
file_SLAMS_data = file_SLAMS_data.sort_values( by= 'data_parsed' )
file_SLAMS_data

```

```python
ALL_HARD_SLAMS = [file_partite_data, file_SLAMS_data ]
DATA_SLAMS_HARD = pd.concat( ALL_HARD_SLAMS )
DATA_SLAMS_HARD = DATA_SLAMS_HARD.sort_values( by='data_parsed' )

DATA_SLAMS_HARD
```

```python
LISTA_COMPLETA_PARTITE = list( DATA_SLAMS_HARD['PARTITE'] )
print( len( LISTA_COMPLETA_PARTITE ) )
print( LISTA_COMPLETA_PARTITE )

lista_indici_slams = []
lista_partite_slam = []

for idx, v in enumerate ( LISTA_COMPLETA_PARTITE ):
    if 'CLAY' in v :
        lista_indici_slams.append( idx )
        lista_partite_slam.append( v )
        
print( len( lista_indici_slams ) )
print( lista_indici_slams )
print( lista_partite_slam )
```

```python
LISTONE_INDICI_LAST_1 = []
LISTONE_PARTITE_LAST_1 = []

LISTE_ANNIDATE_INDICI = []
LISTE_ANNIDATE_PARTITE = []

partite = list( data_indici['PARTITE'] )

for idx, v in enumerate ( data_indici.INDICI ):
    data_slam_nido = data_slam_idx.loc[ data_slam_idx.IDX<v ]
    LISTE_ANNIDATE_PARTITE = list( data_slam_nido.PARTITE_SLAM )
    LISTE_ANNIDATE_PARTITE_LAST_1 = LISTE_ANNIDATE_PARTITE[-120:]
    LISTONE_PARTITE_LAST_1.append( LISTE_ANNIDATE_PARTITE_LAST_1 )
    #LISTONE_PARTITE_LAST_1.append( LISTE_ANNIDATE_PARTITE_LAST_1 )

#print( LISTONE_PARTITE_LAST_1 )
```

```python
dict_LAST_8_PARTITE = {}
for k,v in zip( partite, LISTONE_PARTITE_LAST_1):
    dict_LAST_8_PARTITE[k]=v
```

```python
DATA_LAST_8 = pd.DataFrame( dict_LAST_8_PARTITE )
DATA_LAST_8 = DATA_LAST_8.T
DATA_LAST_8
```

```python
import re

diffs_LAST_1_list = []
diffs_LAST_2_list = []
diffs_LAST_3_list = []
diffs_LAST_4_list = []
diffs_LAST_6_list = []
diffs_LAST_8_list = []



for k,v in dict_LAST_8_PARTITE.items():
    n1 = k.split("_")[1]
    n1 = '_'+n1+'_'
    #print( n1 )
    n2 = k.split("_")[2]
    n2 = '_'+n2+'_'
    
    ######### ATTENZIONE A LAST_n e a lista[ begin:stop ]
    
    
    ######################### LAST 1 SLAM
    stringa_lista_LAST_1 = ''
    for val in v[-15:]:
        stringa_lista_LAST_1 = stringa_lista_LAST_1+val
    #print( stringa_lista )
    list_matches_string_n1_LAST_1 = len( re.findall(n1 , stringa_lista_LAST_1) )
    list_matches_string_n2_LAST_1 = len( re.findall(n2 , stringa_lista_LAST_1) )
    diffs_LAST_1 = list_matches_string_n1_LAST_1-list_matches_string_n2_LAST_1
    diffs_LAST_1_list.append( diffs_LAST_1 )
    
    
    
    
    ######################### LAST 2 SLAM
    stringa_lista_LAST_2 = ''
    for val in v[-30:]:
        stringa_lista_LAST_2 = stringa_lista_LAST_2+val
    #print( stringa_lista )
    list_matches_string_n1_LAST_2 = len( re.findall(n1 , stringa_lista_LAST_2) )
    list_matches_string_n2_LAST_2 = len( re.findall(n2 , stringa_lista_LAST_2) )
    diffs_LAST_2 = list_matches_string_n1_LAST_2-list_matches_string_n2_LAST_2
    diffs_LAST_2_list.append( diffs_LAST_2 )
    
    
    
    
    ######################### LAST 3 SLAM
    stringa_lista_LAST_3 = ''
    for val in v[-45:]:
        stringa_lista_LAST_3 = stringa_lista_LAST_3+val
    #print( stringa_lista )
    list_matches_string_n1_LAST_3 = len( re.findall(n1 , stringa_lista_LAST_3) )
    list_matches_string_n2_LAST_3 = len( re.findall(n2 , stringa_lista_LAST_3) )
    diffs_LAST_3 = list_matches_string_n1_LAST_3-list_matches_string_n2_LAST_3
    diffs_LAST_3_list.append( diffs_LAST_3 )
    
    
    ######################### LAST 4 SLAM
    stringa_lista_LAST_4 = ''
    for val in v[-60:]:
        stringa_lista_LAST_4 = stringa_lista_LAST_4+val
    #print( stringa_lista )
    list_matches_string_n1_LAST_4 = len( re.findall(n1 , stringa_lista_LAST_4) )
    list_matches_string_n2_LAST_4 = len( re.findall(n2 , stringa_lista_LAST_4) )
    diffs_LAST_4 = list_matches_string_n1_LAST_4-list_matches_string_n2_LAST_4
    diffs_LAST_4_list.append( diffs_LAST_4 )
    
    
    ######################### LAST 6 SLAM
    stringa_lista_LAST_6 = ''
    for val in v[-90:]:
        stringa_lista_LAST_6 = stringa_lista_LAST_6+val
    #print( stringa_lista )
    list_matches_string_n1_LAST_6 = len( re.findall(n1 , stringa_lista_LAST_6) )
    list_matches_string_n2_LAST_6 = len( re.findall(n2 , stringa_lista_LAST_6) )
    diffs_LAST_6 = list_matches_string_n1_LAST_6-list_matches_string_n2_LAST_6
    diffs_LAST_6_list.append( diffs_LAST_6 )
    
    ######################### LAST 8 SLAM
    stringa_lista_LAST_8 = ''
    for val in v[-120:]:
        stringa_lista_LAST_8 = stringa_lista_LAST_8+val
    #print( stringa_lista )
    list_matches_string_n1_LAST_8 = len( re.findall(n1 , stringa_lista_LAST_8) )
    list_matches_string_n2_LAST_8 = len( re.findall(n2 , stringa_lista_LAST_8) )
    diffs_LAST_8 = list_matches_string_n1_LAST_8-list_matches_string_n2_LAST_8
    diffs_LAST_8_list.append( diffs_LAST_8 )
    
    
    
    
    print( list_matches_string_n1_LAST_6,"-", list_matches_string_n2_LAST_6,
          "  diff_LAST_1: ", diffs_LAST_1, "  diff_LAST_2: ", diffs_LAST_2,
          "  diff_LAST_3: ", diffs_LAST_3, "  diff_LAST_4: ", diffs_LAST_4,
          "  diff_LAST_6: ", diffs_LAST_6, "  diff_LAST_8: ", diffs_LAST_8  )
          
        
    
    
    
    
    
    
            

        
    
```
