```python
import pandas as pd
import numpy as np


file_partite = 'data_totale_meno_dubbie_games.xlsx'
file_partite_read = pd.read_excel( file_partite )
file_partite_data = pd.DataFrame( file_partite_read )
file_partite_data


file_date_tornei = 'dataframe_interesse.xlsx'
file_date_tornei_read = pd.read_excel( file_date_tornei )
file_date_tornei_data = pd.DataFrame( file_date_tornei_read )


```

```python
file_partite_data.drop( columns=['unicode'], inplace=True)
file_partite_data
```

```python
file_date_tornei_data
```

```python
data_parsed_partite = []
conto = 0

for idx, v in enumerate ( file_partite_data.PARTITE ):
    for ind, val in enumerate ( file_date_tornei_data.UNICODE_LOCATIONS ):
            locat = v.split("_")[4]
            date_partita = int( v.split("_")[5] )
            if (locat==val) & (date_partita==file_date_tornei_data.ANNO[ind]):
                    data_parsed_partite.append( file_date_tornei_data.data_parsed[ind] )
            

                    
print( len( data_parsed_partite ))            
print( data_parsed_partite )    
```

```python
data_parsed_partite = pd.Series( data_parsed_partite )
conti = data_parsed_partite.value_counts()

for v in conti :
    print( v )
```

```python
file_partite_data['data_parsed'] = data_parsed_partite
file_partite_data = file_partite_data.sort_values( by='data_parsed')

unicode = []
conteggio = 0
for v in file_partite_data.PARTITE:
    conteggio = conteggio+1
    unicode.append( conteggio )
    

file_partite_data['unicode'] = unicode
file_partite_data
```
