```python
partite = list( data_finale.PARTITE )

conto_Zverev = 0
idx_Zverev = []

conto_DEspazio = 0
idx_DEspazio = []

conto_DELspazio = 0
idx_DELspazio = []

conto_trattino = 0
idx_trattino = []



for idx, v in enumerate ( partite ):
    if 'ZVEREV' in v:
        conto_Zverev = conto_Zverev+1
        idx_Zverev.append( idx )
    elif 'DE ' in v:
        conto_DEspazio = conto_DEspazio+1
        idx_DEspazio.append( idx )
    elif 'DEL ' in v:
        conto_DELspazio = conto_DELspazio+1
        idx_DELspazio.append( idx )
    elif '-' in v:
        conto_trattino = conto_trattino+1
        idx_trattino.append( idx )
    
        
        
        
        
        
print("Conto Zverev: ", conto_Zverev )
print("Conto DEspazio: ", conto_DEspazio )
print("Conto DELspazio: ", conto_DELspazio )
print("Conto trattino: ", conto_trattino )
```

```python
indici_da_eliminare = []
for v in idx_Zverev:
    indici_da_eliminare.append( v )
for v in idx_DEspazio:
    indici_da_eliminare.append( v )
for v in idx_DELspazio:
    indici_da_eliminare.append( v )
for v in idx_trattino:
    indici_da_eliminare.append( v )
print( len( indici_da_eliminare ) )
```

```python
data_finale.drop( data_finale.index[indici_da_eliminare], inplace=True )
data_finale
```
