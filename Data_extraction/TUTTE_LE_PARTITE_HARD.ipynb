```python
import pandas as pd
import numpy as np

file_Pune = 'pune_from2002_games.xlsx'
file_Brisbane = 'brisbane_from2002_games.xlsx'
file_Doha = 'doha_from2002_games.xlsx'
file_Sidney = 'sydney_from2002_games.xlsx'
file_Auckland = 'auckland_from2002_games.xlsx'
file_Sofia = 'sofia_from2002_games.xlsx'
file_Montpellier = 'montpellier_from2002_games.xlsx'
file_NewYork = 'newyork_from2002_games.xlsx'
file_Rotterdam = 'rotterdam_from2002_games.xlsx'
file_Delray = 'delray_from2002_games.xlsx'
file_Marsiglia = 'marseille_from2002_games.xlsx'
file_Acapulco = 'acapulco_from2002_games.xlsx'
file_Dubai = 'dubai_from2002_games.xlsx'
file_IndianWells = 'indianwells_from2002_games.xlsx'
file_Miami = 'miami_from2002_games.xlsx'
file_Atlanta = 'atlanta_from2002_games.xlsx'
file_Washington = 'washington_from2002_games.xlsx'
file_LosCabos = 'loscabos_from2002_games.xlsx'
file_Montreal = 'montreal_from2002_games.xlsx'
file_Cincinnati = 'cincinnati_from2002_games.xlsx'
file_WinstonSalem = 'winstonsalem_from2002_games.xlsx'
file_Metz = 'metz_from2002_games.xlsx'
file_SanPietroburgo = 'sanpietroburgo_from2002_games.xlsx'
file_Zhuhai = 'zhuhai_from2002_games.xlsx'
file_Chengdu = 'chengdu_from2002_games.xlsx'
file_Tokyo = 'tokyo_from2002_games.xlsx'
file_Pechino = 'pechino_from2002_games.xlsx'
file_Shanghai = 'shanghai_from2002_games.xlsx'
file_Stoccolma = 'stoccolma_from2002_games.xlsx'
file_Antwerp = 'antwerp_from2002_games.xlsx'
file_Mosca = 'mosca_from2002_games.xlsx'
file_Basilea = 'basel_from2002_games.xlsx'
file_Vienna = 'vienna_from2002_games.xlsx'
file_Parigi = 'parigi_from2002_games.xlsx'
```

```python
lista_file_aggiunti = [file_Pune, file_Brisbane, file_Doha, file_Sidney, file_Auckland,
                       file_Sofia, file_Montpellier, file_NewYork,
                       file_Rotterdam, file_Delray, file_Marsiglia, file_Acapulco,
                       file_Dubai, file_IndianWells, file_Miami, file_Atlanta, file_Washington,
                       file_LosCabos, file_Montreal, file_Cincinnati, file_WinstonSalem,
                       file_Metz, file_SanPietroburgo, file_Zhuhai, file_Chengdu, file_Tokyo,
                       file_Pechino, file_Shanghai, file_Stoccolma, file_Antwerp, file_Mosca,
                       file_Basilea, file_Vienna, file_Parigi ]

lista_dataframes = []

for f in lista_file_aggiunti:
    data_agg = pd.read_excel( f )
    data_agg = pd.DataFrame( data_agg )
    lista_dataframes.append( data_agg )
    
data_totale = pd.concat( lista_dataframes )
data_totale
```

```python
unicode = []
conto = -1
for idx, v in enumerate ( data_totale.PARTITE ):
    conto = conto+1
    unicode.append( conto )

data_totale['unicode'] = unicode 
data_totale = data_totale.set_index("unicode")
data_totale = data_totale.iloc[:,1:]
data_totale
```

```python
data_totale_meno_dubbie = data_totale
data_totale_meno_dubbie
```

```python
data_totale_meno_dubbie.to_excel("data_totale_meno_dubbie_games.xlsx")
```
