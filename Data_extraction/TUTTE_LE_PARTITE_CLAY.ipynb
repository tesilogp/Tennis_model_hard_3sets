```python
import pandas as pd
import numpy as np

file_MonteCarlo = 'montecarlo_from2002.xlsx'
file_Barcelona = 'barcelona_from2002.xlsx'
file_Madrid = 'madrid_from2002.xlsx'
file_Rome = 'rome_from2002.xlsx'
```

```python
lista_file_aggiunti = [file_MonteCarlo, file_Barcelona,
                       file_Madrid, file_Rome]

lista_dataframes = []

for f in lista_file_aggiunti:
    data_agg = pd.read_excel( f )
    data_agg = pd.DataFrame( data_agg )
    lista_dataframes.append( data_agg )
    
data_totale = pd.concat( lista_dataframes )
data_totale
```

```python
unicode = []
conto = -1
for idx, v in enumerate ( data_totale.PARTITE ):
    conto = conto+1
    unicode.append( conto )

data_totale['unicode'] = unicode 
data_totale = data_totale.set_index("unicode")
data_totale = data_totale.iloc[:,1:]
data_totale
```

```python
data_totale_meno_dubbie = data_totale
data_totale_meno_dubbie
```

```python
data_totale_meno_dubbie.to_excel("data_totale_CLAY.xlsx")
```

```python
lista_posti = []
for v in data_totale_meno_dubbie.PARTITE:
    s = v.split("_")[4]
    lista_posti.append( s )
    
lista_posti = pd.Series( lista_posti )
print( lista_posti.unique() )
```
