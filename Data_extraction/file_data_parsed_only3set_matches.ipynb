```python
import pandas as pd
import numpy as np


file_hard = 'file_partite_data_parsed_games.xlsx'
file_clay = 'file_partite_data_parsed_CLAY.xlsx'
file_grass = 'file_partite_data_parsed_GRASS_NOgames.xlsx'



file_hard_read = pd.read_excel( file_hard )
file_clay_read = pd.read_excel( file_clay )
file_grass_read = pd.read_excel( file_grass )


file_hard_data = pd.DataFrame( file_hard_read )
file_clay_data = pd.DataFrame( file_clay_read )
file_grass_data = pd.DataFrame( file_grass_read )

```

```python
frms = [ file_hard_data, file_clay_data, file_grass_data ]
data_all_matches_data_parsed = pd.concat( frms )
data_all_matches_data_parsed = data_all_matches_data_parsed.sort_values( by=['data_parsed'] )
data_all_matches_data_parsed
```
