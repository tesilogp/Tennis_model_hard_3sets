```python
import pandas as pd
import numpy as np


file_partite = 'ALL_MATCHESNO_FILTRO.xlsx'
file_partite_read = pd.read_excel( file_partite )
file_partite_data = pd.DataFrame( file_partite_read )
file_partite_data
```

```python
file_2011_in_poi = file_partite_data.loc[ file_partite_data['data_parsed']>("2010-12-31") ]
file_2011_in_poi.to_excel("file_2011_in_poi.xlsx")
```

```python
locations = []
yrss = []

for v in file_partite_data.PARTITE:
    stringa = v.split("_")[4]
    yr = v.split("_")[5]
    yr = int( yr )
    yrss.append( yr )
    locations.append( stringa )
    
file_partite_data['LOCS'] = locations
file_partite_data['YEAR'] = yrss
    
locations = pd.Series( locations )
locations_uniques = list( locations.unique() )
print( locations_uniques  )
    
```

```python
file_partite_data
```

```python
lista_excel_singoli_tornei_matches = []
lista_locs = []

for v in locations_uniques:
    stringa = v+"_matches.xlsx"
    lista_excel_singoli_tornei_matches.append( stringa )
    lista_locs.append( v )
    data = file_partite_data.loc[ file_partite_data['LOCS']==v ]
    data.to_excel( stringa )
    
```

```python
data_excel_locs = pd.DataFrame()
data_excel_locs['LOCS'] = lista_locs
data_excel_locs['EXCEL_MATCHES'] = lista_excel_singoli_tornei_matches
data_excel_locs.to_excel("data_excel_locs.xlsx")
```

```python
def chunks(lst, n):
    for i in range(0, len(lst), n):
        yield lst[i:i + n]


for nome in locations_uniques:
    stringa_excel_tornei_storico = nome+'_STORICO.xlsx'
    for idx, v in enumerate ( data_excel_locs.EXCEL_MATCHES ):
        if data_excel_locs.LOCS[idx]==nome:
            file_corretto = v
            file_corretto_read = pd.read_excel( file_corretto )
            file_corretto_data = pd.DataFrame( file_corretto_read )
            years_torneo = []
            for ans in file_corretto_data.PARTITE:
                years_torneo.append( int( ans.split("_")[5] ) )
            years_torneo = pd.Series( years_torneo )
            years_torneo_uniques = list( years_torneo.unique() )
        
            n_ott = []
        
        
            for an in years_torneo_uniques:
                file_corretto_data_2010 = file_corretto_data.loc[ ( file_corretto_data['YEAR']==an ) & 
                                                         ( file_corretto_data['segno']==1 ) ]
                for idx, v in enumerate ( file_corretto_data_2010.PARTITE ):
                    if v.split("_")[0]=='F':
                        n_ott.append( v.split("_")[1] )
                        n_ott.append( v.split("_")[2] )
                for idx, v in enumerate ( file_corretto_data_2010.PARTITE ):
                    if v.split("_")[0]=='SF':
                        n_ott.append( v.split("_")[2] )
                for idx, v in enumerate ( file_corretto_data_2010.PARTITE ):
                    if v.split("_")[0]=='QRT':
                        n_ott.append( v.split("_")[2] )
                for idx, v in enumerate ( file_corretto_data_2010.PARTITE ):
                    if v.split("_")[0]=='OTT':
                        n_ott.append( v.split("_")[2] )
    listone = list( chunks( n_ott, 16) )
    zip_torneo = {}
    for k,l in zip( years_torneo_uniques, listone ):
        zip_torneo[k] = l
    dataframe_torneo = pd.DataFrame( zip_torneo )
    dataframe_torneo = dataframe_torneo.T
    dataframe_torneo.to_excel( stringa_excel_tornei_storico )
    dataframe_torneo
            
        
```
