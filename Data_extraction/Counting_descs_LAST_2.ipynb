```python
import pandas as pd
import numpy as np

file_desc = 'DESCRITTORE_STORICO_games/dataframe_unitario_STORICO_games.xlsx'
file_desc_read = pd.read_excel( file_desc )
file_desc_data = pd.DataFrame( file_desc_read )
file_desc_data
```

```python
import re

lista_colonne_data_originale = file_desc_data.columns

diff_QRT_1_dict = {}
for idx, v in enumerate ( lista_colonne_data_originale ):
    if "2_diff_QRT" in v:
        diff_QRT_1_dict[idx]=v
        
print( diff_QRT_1_dict )

data_desc_1_diff_QRT = file_desc_data.iloc[:,list( diff_QRT_1_dict.keys() ) ]
count_1_QRT = data_desc_1_diff_QRT.sum( axis=1 )
print( len( count_1_QRT ) )
print( list( count_1_QRT ) )
        
```

```python
diff_OTT_1_dict = {}
for idx, v in enumerate ( lista_colonne_data_originale ):
    if "2_diff_OTT" in v:
        diff_OTT_1_dict[idx]=v
        
print( diff_OTT_1_dict )

data_desc_1_diff_OTT = file_desc_data.iloc[:,list( diff_OTT_1_dict.keys() ) ]
count_1_OTT = data_desc_1_diff_OTT.sum( axis=1 )
print( len( count_1_OTT ) )
print( list( count_1_OTT ) )
```

```python
desc_PESATO_LAST_1 = []

for idx, v in enumerate ( count_1_OTT ):
    count_pesato = v+( 2*count_1_QRT[idx] )+( 3*count_1_SF[idx] )+( 4*count_1_F[idx] )+( 5*count_1_W[idx] ) 
    desc_PESATO_LAST_1.append( count_pesato )
    
print( desc_PESATO_LAST_1 )
```

```python
desc_count_LAST_2 = pd.DataFrame()
desc_count_LAST_2['PARTITE'] = list( file_desc_data.PARTITE )
desc_count_LAST_2['count_2_OTT'] = count_1_OTT
desc_count_LAST_2['count_2_QRT'] = count_1_QRT
desc_count_LAST_2['count_2_SF'] = count_1_SF
desc_count_LAST_2['count_2_F'] = count_1_F
desc_count_LAST_2['count_2_W'] = count_1_W
desc_count_LAST_2['count_2_pesato'] = desc_PESATO_LAST_1
desc_count_LAST_2
```
