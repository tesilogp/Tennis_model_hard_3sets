```python
lista_interesse = ['Pune','Brisbane','Doha','Sydney','Auckland',
                   'Sofia','Montpellier','New York','Rotterdam',
                   'Delray Beach','Marseille','Acapulco','Dubai',
                   'Indian Wells','Miami','Atlanta','Washington',
                   'Los Cabos','Montreal - Masters','Cincinnati',
                   'Winston Salem','Metz','Saint Petersburg','Zhuhai',
                   'Chengdu','Tokyo (Japan Open)','Beijing','Shanghai - Masters',
                   'Stockholm','Antwerp','Moscow','Basel','Vienna',
                   'Paris - Masters']

```

```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re

url_base = 'https://www.tennisexplorer.com/calendar/atp-men/'

anni = range( 2002, 2022 )
url_totali = []
for a in anni:
    url_tot = url_base+str( a )+'/'
    url_totali.append( url_tot )
    
print( url_totali )

tornei_dict = {}

date_tornei = []
names_list = []
anni_list = []



for u in url_totali:
    r = requests.get( u )
    soup = BeautifulSoup( r.content, 'html.parser' )
    names_find = soup.find_all('th', {'class':'t-name'} )
    year = int( u[-5:-1] )
    for v in names_find:
        names_list.append( v.get_text() )
        anni_list.append( year )
    
   

    dates_f_all = soup.find_all('td', {'class':re.compile('first shortdate')} )
    for val in dates_f_all:
        date_tornei.append( val.get_text() )
    
#print( len( tornei_dict ) )
print( names_list )
print( anni_list )
print( date_tornei )
#print( tornei_dict )
```

```python
indici_tornei_interesse = []

for idx, v in enumerate ( names_list ):
    if v in lista_interesse:
        indici_tornei_interesse.append( idx )
        
print( indici_tornei_interesse )
        
```

```python
data_tornei = pd.DataFrame()
data_tornei['tornei'] = names_list
data_tornei['ANNO'] = anni_list
data_tornei['data_parsed'] = date_tornei

data_tornei
```

```python
dataframe_interesse = data_tornei.iloc[indici_tornei_interesse,:]
dataframe_interesse
```

```python
dataframe_interesse['data_parsed'] = pd.to_datetime(
    dataframe_interesse['data_parsed'], format="%d.%m.%Y")

dataframe_interesse
```

```python
dataframe_interesse = dataframe_interesse.sort_values(by='data_parsed')
dataframe_interesse
```
