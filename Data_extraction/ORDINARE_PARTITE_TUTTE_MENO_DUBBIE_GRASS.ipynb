```python
import pandas as pd
import numpy as np


file_partite = 'data_totale_GRASS.xlsx'
file_partite_read = pd.read_excel( file_partite )
file_partite_data = pd.DataFrame( file_partite_read )
file_partite_data


file_date_tornei = 'dataframe_interesse_GRASS.xlsx'
file_date_tornei_read = pd.read_excel( file_date_tornei )
file_date_tornei_data = pd.DataFrame( file_date_tornei_read )


```

```python
data_parsed_partite = []
conto = 0

for idx, v in enumerate ( file_partite_data.PARTITE ):
    for ind, val in enumerate ( file_date_tornei_data.UNICODE_LOCATIONS ):
            locat = v.split("_")[4]
            date_partita = v.split("_")[5]
            if locat==val :
                if date_partita in file_date_tornei_data.data[ind]:
                    data_parsed_partite.append( file_date_tornei_data.data_parsed[ind] )
            

                    
print( len( data_parsed_partite ))            
print( data_parsed_partite )    
```

```python
data_parsed_partite = pd.Series( data_parsed_partite )
conti = data_parsed_partite.value_counts()

for v in conti :
    print( v )
```
