```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

desc = 'descrittore_finale_games.xlsx'
desc_read = pd.read_excel( desc )
desc_data = pd.DataFrame( desc_read )
# ATTENZIONE ALLA COLONNA DA ELIMINARE
desc_data = desc_data.iloc[:,2:]
desc_data
```

```python
riga_zero = desc_data.iloc[0,:]
for idx, v in enumerate ( riga_zero ):
    print( idx, type( v ) )
```

```python
target = 'file_target_data_2005_games.xlsx'
target_read = pd.read_excel( target )
target_data = pd.DataFrame( target_read )
#target_data = target_data.loc[ target_data.data_parsed>'2005-01-01' ]
target_data
```

```python
SEGNO = target_data.iloc[:,11]
SEGNO
```

```python
conto_3 = 0
conto_4 = 0
idx_1 = []
idx_2 = []
idx_3 = []
idx_4 = []

for idx, v in enumerate ( SEGNO ) :
    if v==3 :
        conto_3 = conto_3+1
        idx_3.append( idx )
    elif v==4 :
        conto_4 = conto_4+1
        idx_4.append( idx )
    elif v==1 :
        idx_1.append( idx )
    elif v==2 :
        idx_2.append( idx )
        
print("Len 3: ", conto_3 )
print("Len 4: ", conto_4 )
print("Lista idx==1 : ", idx_1 )
print("Lista idx==2 : ", idx_2 )

```

```python
desc_data_1 = desc_data.iloc[ idx_1, :]
desc_data_1
```

```python
desc_data_2 = desc_data.iloc[idx_2, :]
desc_data_2
```

```python
lista_desc = [ desc_data_1, desc_data_2 ]

desc_data_12 = pd.concat( lista_desc )
desc_data_12
```

```python
SEGNO_1 = pd.Series( SEGNO[idx_1] )
SEGNO_2 = pd.Series( SEGNO[idx_2] )
lista_SEGNO = [SEGNO_1, SEGNO_2]
SEGNO_12 = pd.concat( lista_SEGNO )

SEGNO_MOD = []
for v in SEGNO_12 :
    if v==1 :
        SEGNO_MOD.append( 1 )
    elif v==2:
        SEGNO_MOD.append( -1 )

print( len( SEGNO_MOD ) )
print( SEGNO_MOD )
        
```

```python
xtr, xte, ytr, yte = train_test_split( desc_data_12, SEGNO_MOD, test_size=0.30,
                                      random_state=1, shuffle=True, stratify=SEGNO_MOD  )
print( np.shape( xtr ) )
print( np.shape( ytr ) )
ytr_series = pd.Series( ytr )
xtr = pd.DataFrame.to_numpy( xtr )
ytr_series_count = ytr_series.value_counts()
print( ytr_series_count )

print( np.shape( xte ) )
print( np.shape( yte ) )
yte_series = pd.Series( yte )
xte = pd.DataFrame.to_numpy( xte )
yte_series_count = yte_series.value_counts()
print( yte_series_count )
print( yte )
```

```python
from sklearn.ensemble import RandomForestClassifier

estimators = [10,15,20]
max_depth = [9,10, 11, 12, 13, 14, 15]
min_samples_leaf = [8,9,10]
min_samples_split = [8,9,10]
subsample = [ 1.00]

params = []
tr_metrics = []
te_metrics = []

for e in estimators :
    for md in max_depth :
        for msl in min_samples_leaf :
            for mss in min_samples_split :
                for sbs in subsample:
                    print("n_estimators: ", e, "; max_depth: ", md, 
                      "; min_samples_leaf: ", msl, "; min_samples_split: ", mss, "; subsample: ", sbs  )
                    stringa_params = str(e)+"_"+str(md)+"_"+str(msl)+"_"+str(mss)
                    params.append( stringa_params )
                    rdfc = GradientBoostingClassifier( n_estimators=e, max_depth=md, min_samples_leaf=msl,
                              min_samples_split=mss, random_state=1, learning_rate=0.1,
                              subsample=sbs )
                    rdfc.fit( xtr, ytr )
                    Y_Tr_rdfc = rdfc.predict( xtr )
                    Y_Te_rdfc = rdfc.predict( xte )
    
                    conto = 0
                    for idx,v in enumerate ( ytr ) :
                      if v==Y_Tr_rdfc[idx] :
                        conto = conto+1
                    print("Tr: ",(conto/len(ytr))*100 )
                    tr_metrics.append((conto/len(ytr))*100)
    
                    conto = 0
                    for idx,v in enumerate ( yte ) :
                      if v==Y_Te_rdfc[idx] :
                        conto = conto+1
                    print("Te: ",(conto/len(yte))*100 )
                    te_metrics.append((conto/len(yte))*100)
                    print( confusion_matrix( yte, Y_Te_rdfc ) )
```

```python
data_ottimizzazione = pd.DataFrame()
data_ottimizzazione['params'] = params
data_ottimizzazione['Tr'] = tr_metrics
data_ottimizzazione['Te'] = te_metrics
data_ottimizzazione
```
